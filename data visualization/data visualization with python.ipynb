```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
# Compare the population growth in the US and China
```

```python
us = data[data.country == "United States"]
```

```python
us
```

```python
# Gives true when the row is US and gives false when the row is US
data.country == "United States"
```

```python
data[data.country == "United States"]
```

```python
china = data[data.country == "China"]
```

```python
china
```

```python
#plotting the population of the us
plt.plot(us.year, us.population)
plt.show
```

```python
#plotting the population of the us. (le8) above represents decimal points in billions. change dec point on y axis to mil
plt.plot(us.year, us.population / 10**6)
plt.show
```

```python
#plotting the population of the china in millions /10**6
plt.plot(china.year, china.population)
plt.show
```

```python
#plotting the population of the china and the us in millions
plt.plot(china.year, china.population / 10**6)
plt.plot(us.year, us.population / 10**6)
plt.show
```

```python
#plotting the population of the china and the us in millions and ADDING A LEGEND LABEL
plt.plot(china.year, china.population / 10**6)
plt.plot(us.year, us.population / 10**6)
plt.legend(["United States", "China"])
plt.xlabel("year")
plt.ylabel("population")
plt.show
```

```python
# Calling the population of the US
us.population
```

```python
# Retrieving the first years population using iloc[]
us.population.iloc[0]
```

```python
# Dividing the whole of the US population by the first years amount [first years pop is shown by 1]
us.population/us.population.iloc[0]
```

```python
## Dividing the whole of the US population by the first years amount and the *100 to get the percentage. 1 shows the first year.
us.population/us.population.iloc[0]*100
```

```python
# FINDING THE PERCENTAGE GROWTH AMOUNT IN EACH COUNTRY
#plotting the population of the china and the us in millions and ADDING A LEGEND LABEL
plt.plot(china.year, china.population/china.population.iloc[0]*100)
plt.plot(us.year, us.population/us.population.iloc[0]*100)
plt.legend(["United States", "China"])
plt.xlabel("year")
plt.ylabel("population growth (first year = 100)")
plt.show
```
